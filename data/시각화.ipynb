```python
df = df.drop_duplicates(subset=['경도','위도'])
```

```python
import folium
```

```python
m = folium.Map(location=[36.479868, 127.289240],
               zoom_start=10, 
               width=750, 
               height=500
              )
m
```

```python
from folium.plugins import MarkerCluster

coords = df[['위도', '경도']]
add = df[['지점명']]


marker_cluster = MarkerCluster().add_to(m)

for lat, long in zip(coords['위도'], coords['경도']):
    folium.Marker([lat, long], icon = folium.Icon(color="red"),tooltip=add).add_to(marker_cluster)
m
```

```python
import pandas as pd
import geopandas as gpd
import matplotlib.pyplot as plt
```

```python
F1 = '/Users/jaeyeon/Desktop/타용/세종특별자치시_버스노선 및 버스정류장20230602/emd_20230729' 
D1 = gpd.read_file(F1, encoding='cp949')
D1.head()

D1.convex_hull.plot(color='gray', edgecolor="w")
```

```python
D1.info()

df1 = D1.astype({ 'EMD_CD' : 'int' })
df1.info()
```

```python
(df1['EMD_CD']< 3611039024) & (3000000000 < df1['EMD_CD'])
```

```python
sejong = df1[(36000000<df1['EMD_CD']) & (df1['EMD_CD']< 40000000)]
sejong
```

```python
ax = sejong.convex_hull.plot(color='pink', edgecolor="w")          
ax.set_axis_off()                            # 축제거
plt.show()
```

```python
print(sejong.crs)                        
sejong.set_crs(epsg=5179, inplace= True)
new_sejong = sejong.to_crs(epsg=4326)
new_sejong
```

```python
new_sejong.to_file('./sejong.geojson' , driver='GeoJSON')
```
